```python
import scipy.sparse as sp
from scipy.sparse import coo_matrix

var1  = coo_matrix(test.SEXOX)
var2  = coo_matrix(test.IDADEX)
var3  = coo_matrix(test.PESOX)
var4  = coo_matrix(test.ALTURAX)
var5  = coo_matrix(test.ANTICOAGULANTE_USOX)
var6  = coo_matrix(test.DMX)
var7  = coo_matrix(test.TABAGISMOX)
var8  = coo_matrix(test.HISTORICO_TEVX)
var9  = coo_matrix(test.TM)
var10 = coo_matrix(test.ASA_RXX)
var11 = coo_matrix(test.IMC)
var12 = coo_matrix(test.result)


vvar1  = coo_matrix(train.SEXOX)
vvar2  = coo_matrix(train.IDADEX)
vvar3  = coo_matrix(train.PESOX)
vvar4  = coo_matrix(train.ALTURAX)
vvar5  = coo_matrix(train.ANTICOAGULANTE_USOX)
vvar6  = coo_matrix(train.DMX)
vvar7  = coo_matrix(train.TABAGISMOX)
vvar8  = coo_matrix(train.HISTORICO_TEVX)
vvar9  = coo_matrix(train.TM)
vvar10 = coo_matrix(train.ASA_RXX)
vvar11 = coo_matrix(train.IMC)
vvar12 = coo_matrix(train.result)

combined_test    = sp.hstack([var1.T, var2.T, var3.T,var4.T,var5.T,var6.T,var7.T,var8.T,var9.T,var10.T,var11.T], format='csr')
combined_train   = sp.hstack([vvar1.T, vvar2.T, vvar3.T,vvar4.T,vvar5.T,vvar6.T,vvar7.T,vvar8.T,vvar9.T,vvar10.T,vvar11.T], format='csr')
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, random_state=12345)

from imblearn.pipeline import make_pipeline
from imblearn.over_sampling import RandomOverSampler
rf_pipeline = make_pipeline(RandomOverSampler(random_state=777),rf)
rf_model  = rf_pipeline.fit(combined_train, train.result)

importances = rf.feature_importances_
print(importances)
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, f1_score
def algo_cv(splits, X, Y, pipeline, average_method):
    
    kfold = StratifiedKFold(n_splits=splits, shuffle=True, random_state=777)
    accuracy = []
    precision = []
    recall = []
    f1 = []
    for train, test in kfold.split(X, Y):
        algo_fit = pipeline.fit(X[train], Y[train])
        prediction = algo_fit.predict(X[test])
        scores = algo_fit.score(X[test],Y[test])
        
        accuracy.append(scores * 100)
        precision.append(precision_score(Y[test], prediction, average=average_method)*100)
        print('              Event 1     Event 2')
        print('precision:',precision_score(Y[test],    prediction,    average=None))
        recall.append(recall_score(Y[test],    prediction,    average=average_method)*100)
        print('recall:   ',recall_score(Y[test],     prediction, average=None))
        f1.append(f1_score(Y[test], prediction, average=average_method)*100)
        print('f1 score: ',f1_score(Y[test], prediction, average=None))
        print('-'*40)

    print("accuracy: %.2f%% (+/- %.2f%%)" % (np.mean(accuracy), np.std(accuracy)))
    print("precision: %.2f%% (+/- %.2f%%)" % (np.mean(precision), np.std(precision)))
    print("recall: %.2f%% (+/- %.2f%%)" % (np.mean(recall), np.std(recall)))
    print("f1 score: %.2f%% (+/- %.2f%%)" % (np.mean(f1), np.std(f1)))
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import precision_score, recall_score, f1_score
def algo_cv(splits, X, Y, pipeline, average_method):
    
    kfold = StratifiedKFold(n_splits=splits, shuffle=True, random_state=777)
    accuracy = []
    precision = []
    recall = []
    f1 = []
    for train, test in kfold.split(X, Y):
        algo_fit = pipeline.fit(X[train], Y[train])
        prediction = algo_fit.predict(X[test])
        scores = algo_fit.score(X[test],Y[test])
        
        accuracy.append(scores * 100)
        precision.append(precision_score(Y[test], prediction, average=average_method)*100)
        print('              Event 1     Event 2')
        print('precision:',precision_score(Y[test],    prediction,    average=None))
        recall.append(recall_score(Y[test],    prediction,    average=average_method)*100)
        print('recall:   ',recall_score(Y[test],     prediction, average=None))
        f1.append(f1_score(Y[test], prediction, average=average_method)*100)
        print('f1 score: ',f1_score(Y[test], prediction, average=None))
        print('-'*40)

    print("accuracy: %.2f%% (+/- %.2f%%)" % (np.mean(accuracy), np.std(accuracy)))
    print("precision: %.2f%% (+/- %.2f%%)" % (np.mean(precision), np.std(precision)))
    print("recall: %.2f%% (+/- %.2f%%)" % (np.mean(recall), np.std(recall)))
    print("f1 score: %.2f%% (+/- %.2f%%)" % (np.mean(f1), np.std(f1)))
```
